```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('41586_2021_3436_MOESM4_ESM.xlsx')
```

```python
df
```

```python
df2=df.drop(0)
```

```python
df2['start_year']=df2['Time period'].map(lambda x:x.split('_')[0])
df2['end_year']=df2['Time period'].map(lambda x:x.split('_')[1])
```

```python
local=df2[df2['Region']!='Global']
local=local[local['Glacier category - tw=tidewater (marine-terminating); ntw=non-tidewater (land-terminating); all=all glaciers']=='all']
```

```python
global_=df2[df2['Region']=='Global']
global_=global_[global_['Glacier category - tw=tidewater (marine-terminating); ntw=non-tidewater (land-terminating); all=all glaciers']=='all']
global_=global_[global_['Time period']!='2000-01-01_2020-01-01']
```

```python
glacier60=local[local['Latitude of region centroid']>=60].copy()
```

```python
glacier60.describe()
```

```python
glacier60[glacier60['Time period']=='2000-01-01_2020-01-01'].sort_values('Mean elevation change rate (m yr-1)').head(20)
```

```python
glacier60[glacier60['Time period']=='2000-01-01_2020-01-01'].sort_values('Mass change rate (Gt yr-1)').head(20)
```

```python
glacier60['year']=glacier60['end_year'].str[:4]
```

```python
glacier60
```

```python
global_['year']=global_['end_year'].str[:4]
```

```python
concated = pd.concat([glacier60,global_])
```

```python
cleaned=concated[concated['Time period']!='2000-01-01_2020-01-01']
```

```python
cleaned['Mean elevation change rate (m yr-1)']=cleaned['Mean elevation change rate (m yr-1)'].astype(float)
cleaned['Mass change rate (Gt yr-1)']=cleaned['Mass change rate (Gt yr-1)'].astype(float)
```

```python
cleaned['elevation_change_rate']=cleaned['Mean elevation change rate (m yr-1)']
cleaned['mass_change_rate']=cleaned['Mass change rate (Gt yr-1)']
```

```python
cleaned[cleaned['Region']=='Global']
```

```python
cleaned['year']=cleaned['year'].astype(int)
```

```python
cleaned.describe()
```

```python
cleaned
```

```python
cleaned2 = cleaned[cleaned['Region']!='Global'].copy()
```

```python
cleaned2
```

```python
Old_list = cleaned2['Region'].unique().tolist()
```

```python
Old_list
```

```python
New_list = ['Alaska', 'Arctic Canada, North', 'Arctic Canada, South',
       'Greenland Periphery', 'Iceland', 'Svalbard and Jan Mayen',
       'Scandinavia', 'Russian Arctic', 'Western Canada', 'Asia, North']
```

```python
df_set = []
for region in New_list:
    loop = cleaned2[cleaned2['Region'].str.contains(region)]
    tempo = loop.groupby('year').agg({'Mean elevation change rate (m yr-1)':'mean'})
    if region == 'Western Canada':
        region = 'Western Canada and USA'
    tempo['Region'] = region
    df_set.append(tempo)
final = pd.concat(df_set,axis=0)
```

```python
final['Region'].unique().tolist()
```

```python
final.to_csv('glacier_height.csv')
```
